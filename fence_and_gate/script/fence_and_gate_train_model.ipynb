```python
import pandas as pd
import re
from html import unescape

# Question label mapping
question_labels = {
    "Which category does the property fall under\\?": "Which category does the property fall under?",
    "What will the project consist of\\?": "What will the project consist of?",
    "What is the intended material for the fence or gate\\?": "What is the intended material for the fence or gate?",
    "What is the desired height of the fence\\?": "What is the desired height of the fence?",
    "What is the estimated length of the fence\\?": "What is the estimated length of the fence?",
    "How many access gates are to be included in the fence design\\?": "How many access gates are to be included in the fence design?",
    "Should the professional be responsible for supplying the materials\\?": "Should the professional be responsible for supplying the materials?",
    "Please indicate any supplementary fence-related services you may need\\?": "Please indicate any supplementary fence-related services you may need?",
    "When would you like the work to start\\?": "When would you like the work to start?",
    "How likely are you to hire a professional\\?": "How likely are you to hire a professional?",
    "Additional details[:]?": "Additional Details"
}

# Compile regex patterns for matching
compiled_patterns = [(re.compile(k, flags=re.IGNORECASE), v) for k, v in question_labels.items()]

# Clean HTML and special characters
def clean_answer(text):
    if not isinstance(text, str):
        return ""
    cleaned = re.sub(r'^[^\w\d]*', '', text).strip()
    cleaned = unescape(cleaned)
    return cleaned

# Extract structured answers only
def extract_answers(text):
    if not isinstance(text, str):
        return pd.Series({v: "" for _, v in compiled_patterns})

    extracted = {v: "" for _, v in compiled_patterns}

    # Find all matches of questions in the text
    matches = list(re.finditer(r'(' + '|'.join(question_labels.keys()) + ')', text, flags=re.IGNORECASE))

    for i, match in enumerate(matches):
        question_text = match.group()
        start = match.end()
        end = matches[i+1].start() if i+1 < len(matches) else len(text)
        answer_text = text[start:end].strip()

        for pattern, label in compiled_patterns:
            if pattern.fullmatch(question_text.strip()):
                cleaned = clean_answer(answer_text)
                if not extracted[label]:
                    extracted[label] = cleaned
                break

    return pd.Series(extracted)

# Apply to DataFrame
result_df = Data['Que/Ans'].apply(extract_answers)

# View result
result_df
```

```python
# Load dataset
df = Data

# Define target variable
target = "Credits"

# Define features
features = [
       'Urgent', 'High', 'Verified', 'Frequent',
       'Which category does the property fall under?', 
       'What will the project consist of?',
       'What is the intended material for the fence or gate?',
       'What is the desired height of the fence?',
       'What is the estimated length of the fence?',
       'How many access gates are to be included in the fence design?',
       'Should the professional be responsible for supplying the materials?',
       'Please indicate any supplementary fence-related services you may need?',
       'When would you like the work to start?',
       'How likely are you to hire a professional?']
```

```python
# Drop rows with missing target values
df = df.dropna(subset=[target])
```

```python

# Identify categorical and numerical features
cat_features = df[features].select_dtypes(include=['object']).columns.tolist()
num_features = df[features].select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python

# Fill missing values
df[cat_features] = df[cat_features].fillna("Unknown")
df[num_features] = df[num_features].fillna(df[num_features].median())
```

```python

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Convert to CatBoost Pool format
train_pool = Pool(X_train, label=y_train, cat_features=cat_features)
test_pool = Pool(X_test, label=y_test, cat_features=cat_features)
```

```python

# Initialize CatBoost model
model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=8,
    loss_function='RMSE',
    cat_features=cat_features,
    verbose=100
)
```

```python
# Train model
model.fit(train_pool, eval_set=test_pool, early_stopping_rounds=50)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python

# Evaluate model
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
```

```python
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")
```
