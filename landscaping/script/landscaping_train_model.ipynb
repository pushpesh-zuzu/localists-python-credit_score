```python
import pandas as pd
import re

# Assuming your DataFrame is already loaded and has the column 'Que/Ans'
# Example: Data = pd.read_csv('your_file.csv')

# Define all possible question patterns and their desired column headers
questions = {
    "What kind of landscaping help are you looking for\\?": "What kind of landscaping help are you looking for?",
    "How big is the area you want landscaped\\?": "How big is the area you want landscaped?",
    "Do you already have a plan for what you want done\\?": "Do you already have a plan for what you want done?",
    "Will you need maintenance after the work is complete\\?": "Will you need maintenance after the work is complete?",
    "How likely are you to hire a professional\\?": "How likely are you to hire a professional?",
    "Additional details[:]?": "Additional Details"
}

# Precompile full question matching pattern
question_regex = "(" + "|".join(questions.keys()) + ")"
compiled_pattern = re.compile(question_regex, flags=re.IGNORECASE)

# Clean answer text
def clean_answer(text):
    return re.sub(r'^[^\w\d]*', '', text).strip() if isinstance(text, str) else text

# Extractor function
def extract_answers(row_text):
    if not isinstance(row_text, str):
        return pd.Series({label: None for label in questions.values()})
    
    parts = compiled_pattern.split(row_text)
    parts = [p for p in parts if p and p.strip()]

    extracted = {label: None for label in questions.values()}

    i = 0
    while i < len(parts) - 1:
        question_match = parts[i].strip()
        answer_text = parts[i + 1].strip() if i + 1 < len(parts) else ''
        for q_pattern, q_label in questions.items():
            if re.fullmatch(q_pattern, question_match, flags=re.IGNORECASE):
                if not extracted[q_label]:
                    extracted[q_label] = clean_answer(answer_text)
        i += 2

    return pd.Series(extracted)

# Apply to dataframe
result_df = Data['Que/Ans'].apply(extract_answers)

# View result
result_df
```

```python
Data = pd.concat([Data, result_df], axis=1)
```

```python
Datak = Data
columns_to_drop = ['S.no', 'Response','Country','Additional Details',
    'Initial', 'Name','Time', 'Badge 1','Additional','Credits',
    'Badge 2', 'Badge 3', 'Badge 4', 'Category', 'Que/Ans'
]

# Drop the columns (if they exist in the DataFrame)
Datak = Datak.drop(columns=[col for col in columns_to_drop if col in Datak.columns])

Datak.to_csv("../data/sample_data_for_prediction_landscaping.csv", index=False)
```

```python
columns_to_drop = ['S.no', 'Lead Id', 'Response','Country','Additional Details',
    'Initial', 'Name','Time', 'Badge 1','Additional',
    'Badge 2', 'Badge 3', 'Badge 4', 'Category', 'Que/Ans'
]

# Drop the columns (if they exist in the DataFrame)
Data = Data.drop(columns=[col for col in columns_to_drop if col in Data.columns])
```

```python
Data.columns
```

```python
# Load dataset
df = Data

# Define target variable
target = "Credits"

# Define features
features = [
   'Location', 'Urgent', 'High', 'Verified', 'Frequent',
       'What kind of landscaping help are you looking for?',
       'How big is the area you want landscaped?', 
       'Do you already have a plan for what you want done?',
       'Will you need maintenance after the work is complete?',
       'How likely are you to hire a professional?']
```

```python
# Drop rows with missing target values
df = df.dropna(subset=[target])
```

```python

# Identify categorical and numerical features
cat_features = df[features].select_dtypes(include=['object']).columns.tolist()
num_features = df[features].select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python

# Fill missing values
df[cat_features] = df[cat_features].fillna("Unknown")
df[num_features] = df[num_features].fillna(df[num_features].median())
```

```python

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Convert to CatBoost Pool format
train_pool = Pool(X_train, label=y_train, cat_features=cat_features)
test_pool = Pool(X_test, label=y_test, cat_features=cat_features)
```

```python

# Initialize CatBoost model
model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=8,
    loss_function='RMSE',
    cat_features=cat_features,
    verbose=100
)
```

```python
# Train model
model.fit(train_pool, eval_set=test_pool, early_stopping_rounds=50)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python

# Evaluate model
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
```

```python
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")
```
